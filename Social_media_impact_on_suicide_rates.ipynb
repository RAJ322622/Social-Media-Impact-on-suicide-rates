```python
New_df=New_df.rename(columns={'Suicide Rate % change since 2010':'Suicide_Rate_2010','Twitter user count % change since 2010':'Twitter_user_2010','Facebook user count % change since 2010':'Facebook_user_2010'})
```

```python
New_df.columns
```

# EDA

```python

# Count occurrences of each 'Suicide_Rate_2010' grouped by gender
Suicide_Rate_gender = New_df.groupby(['sex', 'Suicide_Rate_2010']).size().unstack(fill_value=0)

# Create a bar plot for both genders
plt.figure(figsize=(15, 5))
Suicide_Rate_gender.T.plot(kind='bar', width=0.8, color=['skyblue', 'salmon'], figsize=(15, 5))

# Set labels and title
plt.xlabel('Suicide Rate 2010')
plt.ylabel('Frequency')
plt.title('Distribution of Suicide Rate 2010 by Gender')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.legend(title='Sex')
plt.show()
```

```python

# Count occurrences of each 'Suicide_Rate_2010' grouped by gender
Suicide_Rate_gender = New_df.groupby(['sex', 'Twitter_user_2010']).size().unstack(fill_value=0)

# Create a bar plot for both genders
plt.figure(figsize=(15, 5))
Suicide_Rate_gender.T.plot(kind='bar', width=0.8, color=['skyblue','pink', 'salmon'], figsize=(15, 5))

# Set labels and title
plt.xlabel('Twitter_user_2010')
plt.ylabel('Frequency')
plt.title('Distribution of Twitter_user_2010 by Gender')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.legend(title='Sex')
plt.show()
```

```python

# Count occurrences of each 'Suicide_Rate_2010' grouped by gender
Suicide_Rate_gender = New_df.groupby(['sex', 'Facebook_user_2010']).size().unstack(fill_value=0)

# Create a bar plot for both genders
plt.figure(figsize=(15, 5))
Suicide_Rate_gender.T.plot(kind='bar', width=0.8, color=['skyblue','pink', 'salmon'], figsize=(15, 5))

# Set labels and title
plt.xlabel('Facebook_user_2010')
plt.ylabel('Frequency')
plt.title('Distribution of Facebook_user_2010 by Gender')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.legend(title='Sex')
plt.show()
```
